```python
import cv2
import os
from skimage.metrics import structural_similarity as ssim
folder = 'C:\\Users\\Bhagwan\\Downloads\\New folder'

def load_images_from_folder(folder):
    dic = {}
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            dic[filename] = img
    return dic
```

```python
dic = load_images_from_folder(folder)
```

```python
l= []
final = {}
for i in dic:
    m = []
    for j in dic:
        if i!=j:
            print(i,j,ssim(dic[i], dic[j],multichannel = True))
            m.append([ssim(dic[i], dic[j],multichannel = True),j])
    l.append(i)
    final[i] = m
```

```python
for i in final:
    lis = final[i]
    lis.sort(reverse = True)
    final[i]= lis
```

```python
for i in final:
    print(i.split('.')[0])
    for j in final[i]:
        print(j[1].split('.')[0],j[0])
    print()
```
